```
iris = datasets.load_iris()
print(list(iris.keys()))
```

```
# To understand logistic regression we will only take 1 feature. 
x = iris["data"][:, 3:]  #Here we are going to take only 3rd colomn of the data
y = (iris["target"] == 2) # Here we are setting target as 2 coz in our data we have taken column 3 and correspondingly we have taken target 2 which means 
# here we wanted to be print if there is some flower called verginica or not, ethier true or false and this is binary classification
print(y)
```

```
# We wanted to print the above output as 0 or 1
y = (iris["target"] == 2).astype(np.int)
print(y)
```

```
print(x)
```

```
from sklearn.linear_model import LogisticRegression
# Train logistic regression classifier
clf = LogisticRegression()
clf.fit(x, y)
print(clf.predict([[1.6]]))  # Here we have tried giving our patel width as 1.6 and it gives output as 0 means its is not verginica
print(clf.predict([[2.6]]))  # here it is giving as it is verginica
```

```
# Using matplotlip to plot the visualization
# Linspace is a numpy function which is nothing but as we can see down we want 1000 points between 0 and 3
# Reshape means it reashapes the numpy array and down we are creating a 1 d array
x_new = np.linspace(0,3,1000).reshape(-1,1)
print(x_new)
```

```
# Predicts proba(probability) down will tell the actual value of the probability of x_new
y_prob = clf.predict_proba(x_new)
print(y_prob)
```

```
# now we will plot on x axis x_new and on y axis we will plot probabilities y_prob
import matplotlib.pyplot as plt
plt.plot(x_new, y_prob[:,1], label = "verginica") # Here we just want 1 column of y_prob
plt.show()
```
